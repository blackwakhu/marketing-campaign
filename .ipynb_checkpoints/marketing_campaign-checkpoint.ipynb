```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('marketing_campaign.csv',sep='\t')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.NaN, strategy='mean')

df.Income = imp.fit_transform(df.Income[:,np.newaxis])
```

```python
cat_cols = list()
num_cols = list()
for col in df.columns:
    if df[col].nunique() < 15:
        cat_cols.append(col)
    else:
        num_cols.append(col)
for col in cat_cols:
    print(f'{col}: {df[col].unique()}')
```

```python
df1 = df.drop(['Z_Revenue', 'Z_CostContact', 'ID'], axis=1)
df1.head()
```

```python
df1.Dt_Customer = pd.to_datetime(df1.Dt_Customer) 
df1['year'] = df1.Dt_Customer.dt.year
df1['month'] = df1.Dt_Customer.dt.month
df1['day'] = df1.Dt_Customer.dt.day 
df1 = df1.drop(['Dt_Customer'], axis=1)
df1.head()
```

```python
df1.Education.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
labeller = LabelEncoder()

df1.Education = labeller.fit_transform(df1.Education)
df1.Education.value_counts()
```

```python
outlier_list = ['Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases', 'NumWebVisitsMonth']
def tukey(df, col):
    Q1 = np.percentile(df[col], 25, axis=0)
    Q3 = np.percentile(df[col], 75, axis=0)
    IQR = Q3 - Q1
    upper = Q3 + 1.5 *IQR
    lower = Q1 - 1.5 *IQR
    return df[(df[col] < upper) | (df[col] > lower)]
```

```python
for col in outlier_list:
    df1 = tukey(df1, col)
```

```python
plt.scatter(df1.Income, df1.MntFishProducts, c=df1.Response, cmap='viridis')
plt.xlabel('income')
plt.ylabel('fish products')
```

```python
df1 = df[df.Income<600000]
plt.scatter(df1.Income, df1.MntFishProducts, c=df1.Response, cmap='viridis')
plt.xlabel('income')
plt.ylabel('fish products')
```

```python
corrmap = df1.corr()
sns.heatmap(corrmap)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
y = df1.Response
X = df1.drop(['Response'], axis=1)
x_pca = pca.fit_transform(X, y)
plt.scatter(x_pca[:,0],x_pca[:,1], c=y, cmap='plasma')
```
